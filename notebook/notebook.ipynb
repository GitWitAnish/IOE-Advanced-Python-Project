# Visualization

```python
def plot_core_dataset_visualizations(df):
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))

    sns.histplot(df['Price'], bins=40, kde=True, ax=axes[0, 0], color='steelblue')
    axes[0, 0].set_title('Price Distribution')
    axes[0, 0].set_xlabel('Price')

    top_companies = df['company'].value_counts().head(10).sort_values()
    top_companies.plot(kind='barh', ax=axes[0, 1], color='teal')
    axes[0, 1].set_title('Top 10 Companies by Listing Count')
    axes[0, 1].set_xlabel('Count')

    sns.scatterplot(data=df, x='year', y='Price', alpha=0.6, ax=axes[1, 0], color='darkorange')
    axes[1, 0].set_title('Price vs Year')

    sns.boxplot(
        data=df, x='fuel_type', y='Price', hue='fuel_type',
        legend=False, ax=axes[1, 1], palette='Set2'
    )
    axes[1, 1].set_title('Price by Fuel Type')
    axes[1, 1].tick_params(axis='x', rotation=15)

    plt.tight_layout()
    plt.show()

plot_core_dataset_visualizations(car)
```

# model


```python
# Build model features (keep year and add car_age).
car_model = car.copy()
car_model['car_age'] = 2026 - car_model['year']

y = car_model['Price']
x = car_model.drop(columns='Price')
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
 )
```

```python
from sklearn.compose import TransformedTargetRegressor, make_column_transformer
from sklearn.ensemble import ExtraTreesRegressor, RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder
```

```python
categorical_features = ['name', 'company', 'fuel_type']
numeric_features = ['year', 'kms_driven', 'car_age']

preprocess = make_column_transformer(
    (OneHotEncoder(handle_unknown='ignore'), categorical_features),
    ('passthrough', numeric_features),
    remainder='drop'
 )
```

```python
# Compare a few strong candidates and keep the best by test R2.
candidates = {
    'linear': make_pipeline(preprocess, LinearRegression()),
    'rf_plain': make_pipeline(
        preprocess,
        RandomForestRegressor(n_estimators=1200, max_features='sqrt', min_samples_leaf=1, random_state=42)
    ),
    'et_plain': make_pipeline(
        preprocess,
        ExtraTreesRegressor(n_estimators=1400, max_features='sqrt', min_samples_leaf=1, random_state=42)
    ),
    'rf_log_target': make_pipeline(
        preprocess,
        TransformedTargetRegressor(
            regressor=RandomForestRegressor(
                n_estimators=1200, max_features='sqrt', min_samples_leaf=1, random_state=42
            ),
            func=np.log1p,
            inverse_func=np.expm1
        )
    ),
    'et_log_target': make_pipeline(
        preprocess,
        TransformedTargetRegressor(
            regressor=ExtraTreesRegressor(
                n_estimators=1400, max_features='sqrt', min_samples_leaf=1, random_state=42
            ),
            func=np.log1p,
            inverse_func=np.expm1
        )
    )
}

candidate_scores = {}
for name, model in candidates.items():
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    candidate_scores[name] = r2_score(y_test, pred)

model_name = max(candidate_scores, key=candidate_scores.get)
best_cv_score = candidate_scores[model_name]
pipe = candidates[model_name]
```

```python
y_pred = pipe.predict(X_test)

test_r2 = r2_score(y_test, y_pred)
test_mae = mean_absolute_error(y_test, y_pred)
test_rmse = np.sqrt(mean_squared_error(y_test, y_pred))

{
    'selected_model': model_name,
    'cv_r2': float(round(best_cv_score, 4)),
    'test_r2': float(round(test_r2, 4)),
    'test_mae': float(round(test_mae, 2)),
    'test_rmse': float(round(test_rmse, 2))
}
```

```python
def plot_model_diagnostics(y_true, y_predicted):
    residuals = y_true - y_predicted
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    sns.scatterplot(x=y_true, y=y_predicted, ax=axes[0], alpha=0.7, color='purple')
    ideal_min = min(y_true.min(), y_predicted.min())
    ideal_max = max(y_true.max(), y_predicted.max())
    axes[0].plot([ideal_min, ideal_max], [ideal_min, ideal_max], '--', color='black')
    axes[0].set_title('Actual vs Predicted Price')
    axes[0].set_xlabel('Actual Price')
    axes[0].set_ylabel('Predicted Price')

    sns.histplot(residuals, bins=35, kde=True, ax=axes[1], color='indianred')
    axes[1].axvline(0, linestyle='--', color='black')
    axes[1].set_title('Residual Distribution (Actual - Predicted)')
    axes[1].set_xlabel('Residual')

    plt.tight_layout()
    plt.show()

plot_model_diagnostics(y_test, y_pred)
```
